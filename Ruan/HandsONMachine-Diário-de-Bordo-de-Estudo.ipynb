```python
Machine Learning - Hands on - Cap a Cap

Objetivo principal: chegar em transformers

Capítulo 1 - ML Landscape

Utilização da máquina para problemas muito grandes para seres humanos, ou que sejam muito 
voláteis para serem resolvidos, ou com muitas regras, muitos ajustes. Normalmente estamos falando
de usar machine learning para realizar um estudo do problema, e ver quais são as soluções que 
serão encontradas. Também muito utilizado para problemas em que o próprio código possa se adaptar
ao problema (a forma mais conhecida de ser utilizada.), criando assim uma independência da má-
quina quanto ao estudo do problema. 

O capítulo fala de algumas aplicações, e uma delas que pode ser muito interessante fora trans-
formers, é o de anomaly detection, para a área de segurança. Talvez algo envolvendo regras de 
firewall, ou talvez alteração em pacotes? -

Podemos classificar os diversos tipos de sistemas de machine learning  a partir de certas cate-
gorias: 

- Treinamento com ou sem supervisão humana (supervisado, não-supervisado, semi supervisado e 
Reinforcement Learning [RL])

- Aprendizado incremental enquanto está funcionando (online X batch learning)

- Funcionamento a partir da comparação entre pontos de data novos com já existentes, ou então 
detectando padrões na data de treinamento e criando um modelo preditivo (instance-based vs 
model-based learning)

Supervised Learning

O set de treinamento de dados que você alimenta os algoritmos inclui as soluções, que são 
chamados labels(rótulos). 

Uma task comum desse tipo de treinamento é a classificação. Seria no caso a classificação de 
novas informações a partir dos labels inseridos. Outra task comum é a de predição de um valor 
numérico almejado, a partir de um conjunto de features (chamado de preditores). O nome dessa 
task é regressão. 

Unsupervised Learning

A data de treinamento não é rotulada. O sistema tenta aprender sozinho. 
Um exemplo de task é a redução dimensional. Neste, queremos simplificar a data sem perder 
informação. 

Semisupervised Learning

Trabalha com boa parte dos dados sendo unlabeled, e um pouco dos dados labered. A maior parte de algoritmos semi-super
visionados são combinações dos anteriores. Um exemplo bom disto é a deep belief networks (DBN), que é baseado em compo-
nentes não-supervisionados chamados de restrict Boltzman machines (RBMs) empilhadas. Elas são treinadas sequencialmente
de uma forma não supervisionada, e ai todo o sistema é afinado usando técnicas supervisionadas. 

Reinforcement Learning

O agente (sistema que aprende), observa o ambiente, seleciona e faz ações, e ganha recompensas em retorno, ou penaliza-
ções. Aprende por si só qual a melhor política (estratégia) para ganhar mais recompensas overtime. Exemplo é o DeepMinds'
AlphaGo. 

Batch Learning
Aprende a partir de todo o dado disponível, não aprende de forma incremental. Para atualizar este sistema, é necessário
refazer todo o treinamento, com os dados novos e velhos.Um exemplo é o offline learning.

Online Learning

Treina de forma incremental, alimentando o sistema com instâncias de data de forma sequencial, individualmente ou por 
pequenos grupos, chamados de mini-batches. 

Instance-based learning

O sistema aprende os exemplos a risca (ou seja aplica os casos conhecidos, e depois generaliza para casos novos utili-
zando uma métrica de similaridade. 


Model-based learning

Outra forma de generalizar a partir de exemplos é construindo um modelo destes exemplos, e usar esse modelo para criar
predições. 


Sempre analise a big picture dos seus dados, para que você possa encontrar a melhor forma de trabalhar com eles. 
Primeiramente, vamos definir qual será o tipo de modelo a ser utilizado. Depois, quais serão os parâmetros que importa-
rão, o qual depende do modelo escolhido. 
Para escolher quais serão os melhores parâmetros para aquela aplicação, o interessante é definir uma função de utilidade (ou função fitness), que mede o quão bom o seu modelo é, ou então uma cost
function, que define o quão ruim seu modelo é. Daí, entra um algoritmo que é alimentado com os seus exemplos de treino,
e que encontra os melhores parâmetros possíveis. O nome disso é treinamento do modelo. Por fim, inferencia.
                                       
A medida que dei continuidade na leitura, notei que é mais interessante e eficiente, ao final de cada cap, responder
as questões, ao invés de ficar fazendo anotações. Seguirei este modelo. 
                                       
                                     
                                       
                                       


```
